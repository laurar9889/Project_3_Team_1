```python
#split up span in to start and end date
Cleaned_Data['Period']
```

```python
#Splitting years by pattern '-'
Cleaned_Data['Period'].str.split(pat='-')
```

```python
#checking plit where 0=starting year=LHS values
Cleaned_Data['Period'].str.split(pat='-').str[0]
```

```python
#adding left hand values to new column name 'Starting_Year'
Cleaned_Data['Starting_Year'] = Cleaned_Data['Period'].str.split(pat='-').str[0]
```

```python
#checking plit where 1=year_Ending=RHS values
Cleaned_Data['Period'].str.split(pat='-').str[1]
```

```python
##adding Right hand values to new column name 'Year_Ending'
Cleaned_Data['Year_Ending'] = Cleaned_Data['Period'].str.split(pat='-').str[1]
```

```python
Cleaned_Data.head(5)
```

```python
# checking null values for column 'year_Ending'
Cleaned_Data.isnull()
```

```python
#checking records who is carrying null values for year_ending column
Cleaned_Data[Cleaned_Data['Year_Ending'].isna()==1]
```

```python
#replacing null values by zero
Cleaned_Data['Year_Ending'] = Cleaned_Data['Year_Ending'].fillna(0)
Cleaned_Data
```

```python
Cleaned_Data=Cleaned_Data.drop(['Period'],axis=1)
Cleaned_Data.head(5)
```

```python
# Varifying null values for column 'year_Ending'
Cleaned_Data.isnull()
```

```python
#Checking Data type
Cleaned_Data.dtypes
```

```python
#change data type
Cleaned_Data['Starting_Year'] = pd.to_numeric(Cleaned_Data['Starting_Year'], errors='coerce')
Cleaned_Data['Year_Ending'] = pd.to_numeric(Cleaned_Data['Year_Ending'], errors='coerce')
Cleaned_Data['Gender'] = Cleaned_Data['Gender'].astype('category')
```
