Normalization - First Normal Form (1NF)

```python
df.duplicated()
```

Therefor, 
In 1NF, ensure each column contains atomic values, and each record is unique.
> Our table already satisfies 1NF since all columns contain atomic values and records are unique.

Second Normal Form (2NF)

1. Normalize to Indicators table (2NF)

```python
indicators = df[['IndicatorCode', 'Indicator']].drop_duplicates().reset_index(drop=True)
print("Indicators Table:")
indicators
```

```python
indicators.to_csv("Normalized_Tables/Indicators_Table.csv",index=False)
print("indicators_Table extracted under Normalized_Table folder")
```

2. Normalize to Locations table (2NF)

```python
locations = df[['Patents_Continent_Code', 'ParentLocation_Continent']].drop_duplicates().reset_index(drop=True)
print("\nLocations Table:")
locations
```

```python
locations.to_csv("Normalized_Tables/Locations_Table.csv",index=False)
print("Locations_Table extracted under Normalized_Table folder")
```

3. Normalize to Countries table (3NF)

```python
countries = df[['Country_Code', 'Country', 'Patents_Continent_Code']].drop_duplicates().reset_index(drop=True)
print("\nCountries Table:")
countries
```

```python
countries.to_csv("Normalized_Tables/Countries_Table.csv",index=False)
print("Countries_Table extracted under Normalized_Table folder")
```

4. Normalize to Period Range table

```python
period_ranges = df[['Year_Period_Range']].drop_duplicates().reset_index(drop=True)
period_ranges['PeriodRangeID'] = period_ranges.index + 1
period_ranges
```

```python
period_ranges.to_csv("Normalized_Tables/Period_Ranges_Table.csv",index=False)
print("Period_Ranges_Table extracted under Normalized_Table folder")
```

Create the Main Data table (Fact)

```python
main_data = df[['IndicatorCode', 'Country_Code', 'Gender', 'Age_Group', 'Planning_Value_Rate', 'FactComments', 
                'Language_Code', 'Starting_Year', 'Year_Ending', 'Year_Group','Year_Period_Range']].copy()
```

```python
main_data = main_data.merge(period_ranges, on='Year_Period_Range')
main_data = main_data.drop(columns=['Year_Period_Range'])
main_data
```

```python
main_data.to_csv("Normalized_Tables/Main_data_Table.csv",index=False)
print("Main_Data_Table extracted under Normalized_Table folder")
```
